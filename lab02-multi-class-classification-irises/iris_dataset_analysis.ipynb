```python
import pandas as pd
import numpy as np
import tensorflow as tf
import random
import time
import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense, Input
from keras.utils import to_categorical
from keras.backend import clear_session

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split


seed = 42
np.random.seed(seed)
tf.random.set_seed(seed)
random.seed(seed)

dataframe = pd.read_csv("iris_dataset.csv", header=None)
dataset = dataframe.values
X = dataset[:,0:4].astype(float)
Y = dataset[:,4]

encoder = LabelEncoder()
encoder.fit(Y)
encoded_Y = encoder.transform(Y)
dummy_y = to_categorical(encoded_Y)

X_train, X_test, y_train, y_test = train_test_split(
    X, dummy_y,
    test_size=0.2,
    random_state=seed,
    stratify=encoded_Y
)


def build_model(hidden_layers):
    model = Sequential()
    model.add(Input(shape=(4,)))

    for units in hidden_layers:
        model.add(Dense(units, activation='relu'))
        
    model.add(Dense(3, activation='softmax'))

    model.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model


def plot_history(history, title):
    h = history.history

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].plot(h["loss"])
    axes[0].plot(h["val_loss"])
    axes[0].set_title("Loss")
    axes[0].set_xlabel("Epoch")
    axes[0].set_ylabel("Loss")
    axes[0].legend(["train", "val"])

    axes[1].plot(h["accuracy"])
    axes[1].plot(h["val_accuracy"])
    axes[1].set_title("Accuracy")
    axes[1].set_xlabel("Epoch")
    axes[1].set_ylabel("Accuracy")
    axes[1].legend(["train", "val"])

    fig.suptitle(title)
    plt.tight_layout()
    plt.show()


architectures = [
    [4],
    [8],
    [16],
    [32],
    [64],
    [8, 8],
    [16, 8],
    [32, 16],
    [64, 32],
    [64, 32, 16],
]

results_arch = []

for arch in architectures:
    clear_session()

    np.random.seed(seed)
    tf.random.set_seed(seed)
    random.seed(seed)

    model = build_model(arch)

    start = time.time()
    history = model.fit(
        X_train, y_train, 
        batch_size=10, 
        epochs=75, 
        validation_split=0.1
    )
    train_time = time.time() - start

    test_loss, test_acc = model.evaluate(X_test, y_test)

    best_val_acc = max(history.history["val_accuracy"])
    best_val_loss = min(history.history["val_loss"])

    results_arch.append({
        "architecture" : "-".join(map(str, arch)),
        "best_val_accuracy" : best_val_acc,
        "best_val_loss" : best_val_loss,
        "test_accuracy" : test_acc,
        "test_loss" : test_loss,
        "train_time_sec" : train_time
    })

    print(f"\nArchitecture: {arch}")
    print(f"Training time: {train_time:.2f} s")
    print(f"Best val_accuracy: {best_val_acc:.4f}")
    print(f"Best val_loss: {best_val_loss:.4f}")
    print(f"Test accuracy: {test_acc:.4f}")
    print(f"Test loss: {test_loss:.4f}")

    plot_history(history, title=f"Architecture: {arch}\nTraining time: {train_time:.2f} s")

results_arch_df = pd.DataFrame(results_arch)

print("\nСравнительная таблица:")
display(results_arch_df)
```

```python
from keras.optimizers import Adam, SGD, RMSprop, Adagrad, Adadelta


BEST_ARCH = [64, 32, 16]

def build_model_with_optimizer(optimizer, hidden_layers=BEST_ARCH):
    model = Sequential()
    model.add(Input(shape=(4,)))

    for units in hidden_layers:
        model.add(Dense(units, activation='relu'))
    
    model.add(Dense(3, activation='softmax'))

    model.compile(
        optimizer=optimizer,
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model


optimizers = [
    ("Adam (lr=0.001)", Adam(learning_rate=0.001)),
    ("Adam (lr=0.0005)", Adam(learning_rate=0.0005)),

    ("SGD (lr=0.01)", SGD(learning_rate=0.01)),
    ("SGD (lr=0.01 m=0.9)", SGD(learning_rate=0.01, momentum=0.9)),

    ("RMSprop (lr=0.001)", RMSprop(learning_rate=0.001)),
    ("RMSprop (lr=0.0005)", RMSprop(learning_rate=0.0005)),

    ("Adagrad (lr=0.01)", Adagrad(learning_rate=0.01)),
    ("Adagrad (lr=0.005)", Adagrad(learning_rate=0.005)),

    ("Adadelta (default)", Adadelta()),
    ("Adadelta (rho=0.95)", Adadelta(rho=0.95))
]

results_opt = []

for name, opt in optimizers:
    clear_session()

    np.random.seed(seed)
    tf.random.set_seed(seed)
    random.seed(seed)

    model = build_model_with_optimizer(opt)

    start = time.time()
    history = model.fit(
        X_train, y_train,
        batch_size=10,
        epochs=75,
        validation_split=0.1
    )
    train_time = time.time() - start

    test_loss, test_acc = model.evaluate(X_test, y_test)

    best_val_acc = max(history.history["val_accuracy"])
    best_val_loss = min(history.history["val_loss"])

    results_opt.append({
        "optimizer": name,
        "best_val_accuracy": best_val_acc,
        "best_val_loss": best_val_loss,
        "test_accuracy": test_acc,
        "test_loss": test_loss,
        "train_time_sec": train_time
    })

    print(f"\nOptimizer: {name}")
    print(f"Training time: {train_time:.2f} s")
    print(f"Best val_accuracy: {best_val_acc:.4f}")
    print(f"Best val_loss: {best_val_loss:.4f}")
    print(f"Test accuracy: {test_acc:.4f}")
    print(f"Test loss: {test_loss:.4f}")

    plot_history(history, title=f"Optimizer: {name}\nTraining time: {train_time:.2f} s")

results_opt_df = pd.DataFrame(results_opt)

print("\nСравнительная таблица:")
display(results_opt_df)
```

```python
BEST_OPT = lambda: Adam(learning_rate=0.001)

def build_best_model(optimizer_fn=BEST_OPT, hidden_layers=BEST_ARCH):
    model = Sequential()
    model.add(Input(shape=(4,)))

    for units in hidden_layers:
        model.add(Dense(units, activation='relu'))
    
    model.add(Dense(3, activation='softmax'))

    model.compile(
        optimizer=optimizer_fn(),
        loss='categorical_crossentropy',
        metrics=['accuracy']
    )
    return model


fit_experiments = [
    {"epochs": 40,  "batch_size": 5,  "validation_split": 0.1},
    {"epochs": 100,  "batch_size": 5,  "validation_split": 0.1},
    {"epochs": 150, "batch_size": 5,  "validation_split": 0.1},

    {"epochs": 40,  "batch_size": 15, "validation_split": 0.1},
    {"epochs": 100,  "batch_size": 15, "validation_split": 0.1},
    {"epochs": 150, "batch_size": 15, "validation_split": 0.1},

    {"epochs": 40,  "batch_size": 30, "validation_split": 0.1},
    {"epochs": 100,  "batch_size": 30, "validation_split": 0.1},
    {"epochs": 150, "batch_size": 30, "validation_split": 0.1},
]

results_fit = []

for cfg in fit_experiments:
    clear_session()

    np.random.seed(seed)
    tf.random.set_seed(seed)
    random.seed(seed)

    model = build_best_model()

    start = time.time()
    history = model.fit(
        X_train, y_train,
        epochs=cfg["epochs"],
        batch_size=cfg["batch_size"],
        validation_split=cfg["validation_split"]
    )
    train_time = time.time() - start

    test_loss, test_acc = model.evaluate(X_test, y_test)

    best_val_acc = max(history.history["val_accuracy"])
    best_val_loss = min(history.history["val_loss"])

    results_fit.append({
        "epochs": cfg["epochs"],
        "batch_size": cfg["batch_size"],
        "validation_split": cfg["validation_split"],
        "best_val_accuracy": best_val_acc,
        "best_val_loss": best_val_loss,
        "test_accuracy": test_acc,
        "test_loss": test_loss,
        "train_time_sec": train_time
    })

    print("\nFit params:", cfg)
    print(f"Training time: {train_time:.2f} s")
    print(f"Best val_accuracy: {best_val_acc:.4f}")
    print(f"Best val_loss: {best_val_loss:.4f}")
    print(f"Test accuracy: {test_acc:.4f}")
    print(f"Test loss: {test_loss:.4f}")

    plot_history(
        history,
        title = (
            f"Fit: epochs={cfg['epochs']}, batch={cfg['batch_size']}, val={cfg['validation_split']}\n"
            f"Training time: {train_time:.2f} s"
        )
    )

results_fit_df = pd.DataFrame(results_fit)

print("\nСравнительная таблица:")
display(results_fit_df)
```
